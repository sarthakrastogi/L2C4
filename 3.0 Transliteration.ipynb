#Hindi

```python
lang='hi'

romanised_hi_keys = []

for key in tqdm(hi_keys):
    r = ItransTransliterator.to_itrans(key,lang)
    r = r.replace("j~n", "gy").replace(".", "").replace("^", "").replace("ॉ", "o").replace("a़", "on").replace("ऑ", "o").replace("ँ", "n")

    romanised_hi_keys.append(r)
```

```python
hindi_embeddings = {}
count = 0

for key in tqdm(range(len(hi_keys))):
    if key in hindi_embeddings.keys():
        key += 'a'*count
        count += 1

    hindi_embeddings.update({
                            vectors_list[key] : hi_keys[key]
                            })
    

with open(project_path + "Translated_Embeddings/hi_embeddings.pickle", 'wb') as f:
    pickle.dump(hindi_embeddings, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
hindi_embeddings = {}
count = 0

for i in tqdm(range(len(romanised_hi_keys))):
    if romanised_hi_keys[i] in hindi_embeddings.values():
        romanised_hi_keys[i] += 'a'*count
        count += 1

    hindi_embeddings.update({
                            vectors_list[i] : romanised_hi_keys[i]
                            })

with open(project_path + "Transliterated_Embeddings/hi_embeddings.pickle", 'wb') as f:
    pickle.dump(hindi_embeddings, f, protocol=pickle.HIGHEST_PROTOCOL)
```
