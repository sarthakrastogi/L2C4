```python
import re
from tqdm.notebook import tqdm
import pickle
```

```python
with open("Resources/ae.txt", encoding="UTF-8") as file_in:
    lines = []
    for line in file_in:
        lines.append(line.rstrip())
```

```python
lines = lines[lines.index('<L>1<pc>001<k1>a<k2>a'):lines.index('; ABBREVIATIONS')]
```

```python
sa_keys = {}
for line in tqdm(lines):
    if line[:3] == '<L>':
        en_word = re.search('<k1>(.*)<k2>', line).group(1)
        nextline = lines[lines.index(line) + 1]
        sa_words = [item for sublist in [i.split(',') for i in re.findall(r'{#(.*?)\#}', nextline)] for item in sublist]
        if sa_words is None:
            
        sa_keys.update({en_word : sa_words})
```

```python
with open("Translated_Embeddings/en_embeddings.pickle", "rb") as f:
    en_emb = pickle.load(f)
```

```python
sa_emb = {}
for k, v in tqdm(en_emb.items()):
    try:
        trans = sa_keys[v]
    except KeyError:
        trans = None
    sa_emb.update({k : trans})  
        
```

```python
with open("Translated_Embeddings/sa_embeddings.pickle", "wb") as f:
    pickle.dump(sa_emb, f)
```
