```python
all_datasets_url = "https://www.kaggle.com/datasets?sort=votes&fileType=csv&page="
baseurl = "https://www.kaggle.com"
```

```python
num_pages = 500
supposed_num_pages = 2527
```

```python
dataset_links = []
for x in tqdm(range(1,num_pages+1)):

    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    driver.get(all_datasets_url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()
    datasets_on_page = soup.find_all("li",{"class":"sc-fzoaKM hNuJKS"})[14:][:-2] #skipping competitions, datasets, code, etc


    for dataset in datasets_on_page:
        try:
            link = dataset.find("a").get('href')
            dataset_links.append(baseurl + link)
        except AttributeError:
            pass
```
