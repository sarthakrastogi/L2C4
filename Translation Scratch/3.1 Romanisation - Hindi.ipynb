# Romanisation

```python
romanised_hindi_keys = []
```

```python
from indicnlp.transliterate.unicode_transliterate import ItransTransliterator

lang='hi'

for key in tqdm(hindi_keys_list):
    r = ItransTransliterator.to_itrans(key,lang)
    romanised_hindi_keys.append(r)
```

```python
len(romanised_hindi_keys)
```

```python
hindi_embeddings = {}
```

```python
for key in tqdm(range(len(romanised_hindi_keys))):
    hindi_embeddings.update({
                            romanised_hindi_keys[key] : vectors_list[key]
                            })
```

```python
with open(project_path + "Embeddings/hindi_mebeddings.pickle", 'wb') as f:
    pickle.dump(hindi_embeddings, f, protocol=pickle.HIGHEST_PROTOCOL))
```
