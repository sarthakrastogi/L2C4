```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")

def translator(text):
    model_inputs = tokenizer(text, return_tensors="pt")
    generated_tokens = model.generate(**model_inputs,
                                      forced_bos_token_id=tokenizer.lang_code_to_id["ta_IN"])
  
    return tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
translated_words = []
```

```python
len(li[NUMBER])
```

```python
# Translate text
for l in tqdm(li[NUMBER]):
    translated_words.append(translator(l))
```

```python
len(translated_words)
```

```python
translation_table = dict.fromkeys(map(ord, """!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~1234567890qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM"""), None)
```

```python
translated_cleaned_words = []
for word in tqdm(translated_words):
    try:
        translated_cleaned_words.append(word.translate(translation_table).replace("केओआई", "").replace("\u200d", ""))
    except KeyError:
        pass
```
