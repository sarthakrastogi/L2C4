```python
english_keys_list = list(english_embeddings_dict.keys())
```

```python
def printDivisors(n) :
    i = 1
    while i <= n :
        if (n % i==0) :
            print(i,  end = '  ')
        i = i + 1
          
printDivisors(len(english_keys_list)/20)
```

```python
l = np.array_split(english_keys_list, 20)
li = [i.tolist() for i in l]
len(li)
```

```python
LANGUAGE = "german"
NUMBER = 2
```

```python
l0s = np.array_split(li[NUMBER], 879)

l0si = [i.tolist() for i in l0s]
len(l0si)
```

```python
len(l0si[0])
```

```python
from transformers import pipeline

# Init translator
translator = pipeline("translation_en_to_de")

translated_words = []
```

```python
# Translate text
for l in tqdm(l0si):
    translation = translator(l)
    translated_words += translation
```

```python
translated_words
```

```python
translated_words_list  = []
for kvpair in translated_words:
    translated_words_list += kvpair.values()
```

```python
len(translated_words_list)
```

```python
with open(project_path + "Translated_Embeddings/" + str(LANGUAGE) + "_keys_list_" + str(NUMBER) + ".pickle", "wb") as f:
    pickle.dump(translated_words_list, f)#, protocol=pickle.HIGHEST_PROTOCOL)
```
