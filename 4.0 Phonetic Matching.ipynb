```python
def invert_embeddings(emb):
    a = {}
    for k, v in emb.items():
        a.update({v : k})
    return a
```

```python
from Levenshtein import distance 
def match(s1, s2):
    return distance(s1, s2) <= 1
```

```python
def phone_sim_graph(lang1, lang2):
    graph = {}
    emb1, emb2 = load_embeddings(lang1), load_embeddings(lang2)
    emb1_unro, emb2_unro = load_unro_embeddings(lang1), load_unro_embeddings(lang2)
    emb_en = load_embeddings('en')



    if lang2 != 'en':
        for k in tqdm(vectors_list):
            w1, w2 = emb1[k], emb2[k]
            w1_unro, w2_unro = emb1_unro[k], emb2_unro[k]
            if w1_unro != emb_en[k] and w2_unro != emb_en[k]:
                if distance(doublemetaphone(w1)[0], doublemetaphone(w2)[0]) <= 1 and len(w1) > 3:                                               # or
                #if difflib.SequenceMatcher(None, w1, w2).ratio() > 0.98:
                    graph.update({w1 : k})
                    #print(doublemetaphone(w1))

    elif lang2 == 'en':
        for k in tqdm(vectors_list):
            w1, w2 = emb1[k], emb2[k]
            w1_unro, w2_unro = emb1_unro[k], emb2_unro[k]
            if w1_unro != emb_en[k]:
                if distance(doublemetaphone(w1)[0], doublemetaphone(w2)[0]) <= 1 and len(w1) > 3:                                               # or
                #if difflib.SequenceMatcher(None, w1, w2).ratio() > 0.98:
                    graph.update({w1 : k})
                    #print(doublemetaphone(w1))
    return graph
```
